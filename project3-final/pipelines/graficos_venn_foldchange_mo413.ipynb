```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Dados fornecidos
count_A = 306
count_B = 177
count_intersect_AB = 22

# Calcular os valores para o diagrama de Venn
count_only_A = count_A - count_intersect_AB
count_only_B = count_B - count_intersect_AB

# Calcular porcentagens
total = count_only_A + count_only_B + count_intersect_AB
percent_A = (count_only_A / total) * 100
percent_B = (count_only_B / total) * 100
percent_intersect_AB = (count_intersect_AB / total) * 100

# Configurações de cores
color_lightblue = '#ADD8E6'   # Azul claro
color_lightgreen = '#FFB6C1'  # Rosa
color_lightorange = '#FFD700' # Laranja claro

# Plotar o diagrama de Venn
venn = venn2(subsets=(count_only_A, count_only_B, count_intersect_AB),
             set_labels=('Pré-Eclampsia', 'Câncer'),
             set_colors=(color_lightblue, color_lightgreen),
             alpha=0.7)

# Personalizar as áreas do diagrama
venn.get_patch_by_id('10').set_color(color_lightblue)      # A apenas
venn.get_patch_by_id('01').set_color(color_lightgreen)     # B apenas
venn.get_patch_by_id('11').set_color(color_lightorange)    # Interseção

# Adicionar números e porcentagens dentro do diagrama
#venn.get_label_by_id('10').set_text(f'{count_only_A}\n({percent_A:.1f}%)')
#venn.get_label_by_id('01').set_text(f'{count_only_B}\n({percent_B:.1f}%)')
#venn.get_label_by_id('11').set_text(f'{count_intersect_AB}\n({percent_intersect_AB:.1f}%)')

# Mostrar o gráfico
plt.title("Quantidade de proteínas")
plt.show()

# Criar legenda separada
legend_labels = ['Pré-Eclampsia', 'Câncer', 'Compartilhadas']
legend_colors = [color_lightblue, color_lightgreen, color_lightorange]
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in legend_colors]

# Posicionar a legenda no canto direito superior
plt.legend(legend_handles, legend_labels, loc='best')
```
