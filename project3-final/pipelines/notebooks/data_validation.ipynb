```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
exp_path_1 = Path('../../data/raw/proteomica.xlsx')
exp_path_2 = Path('../../data/raw/dados_proteomica_com_foldchange.xlsx')

cancer_sheet = 'Cancer-Rat'
preeclam_sheet = 'Preeclampsia-Rat'
```

```python
cancer_cols = ['Query protein', 'Gene name', 'FC']
rename_cancer_cols = {'Query protein':'protein', 'Gene name':'gene', 'FC':'fc'}
cancer_expression_data_1 = pd.read_excel(exp_path_1, sheet_name=cancer_sheet, usecols=cancer_cols).rename(columns=rename_cancer_cols)
print(cancer_expression_data_1.shape)
cancer_expression_data_1.head(10)
```

```python
cancer_exp_cols = ['Query_protein', 'Gene_name', 'FC']
rename_cancer_exp_cols = {'Query_protein':'protein', 'Gene_name':'gene', 'FC':'fc'}
cancer_expression_data_2 = pd.read_excel(exp_path_2, sheet_name=cancer_sheet, usecols=cancer_exp_cols).rename(columns=rename_cancer_exp_cols)
print(cancer_expression_data_2.shape)
cancer_expression_data_2.head(10)
```

```python
pd.concat([cancer_expression_data_1, cancer_expression_data_2], axis=0, ignore_index=True).drop_duplicates(keep=False)
```

```python
### Pre-eclampsia
```

```python
preeclam_exp_cols = ['protein_firstname', 'Gene name.1', 'Fold change']
rename_preeclam_exp_cols = {'protein_firstname':'protein', 'Gene name.1':'gene', 'Fold change':'fc'}
preeclam_expression_data_1 = pd.read_excel(exp_path_1, sheet_name=preeclam_sheet, usecols=preeclam_exp_cols).rename(columns=rename_preeclam_exp_cols)
preeclam_expression_data_1['gene'] = preeclam_expression_data_1['gene'].str.strip()
print(preeclam_expression_data_1.shape)
preeclam_expression_data_1
```

```python
preeclam_exp_cols = ['protein_firstname', 'Gene_name_correct', 'FC']
rename_preeclam_exp_cols = {'protein_firstname':'protein', 'Gene_name_correct':'gene', 'FC':'fc'}
preeclam_expression_data_2 = pd.read_excel(exp_path_2, sheet_name=preeclam_sheet, usecols=preeclam_exp_cols).rename(columns=rename_preeclam_exp_cols)
preeclam_expression_data_2['gene'] = preeclam_expression_data_2['gene'].str.strip()
print(preeclam_expression_data_2.shape)
preeclam_expression_data_2
```

```python
pd.concat([preeclam_expression_data_1, preeclam_expression_data_2], axis=0, ignore_index=True).drop_duplicates(keep=False)
```
